```python
def hitung_kecepatan_komponen(v0, sudut):
        """Menghitung komponen kecepatan awal"""
        sudut_rad = math.radians(sudut)
        vx = v0 * math.cos(sudut_rad)
        vy = v0 * math.sin(sudut_rad)
        return vx, vy
```

```python
import math

# Inisialisasi variabel-variabel yang dibutuhkan
vx = 10.0   # Kecepatan awal pada sumbu x (contoh nilai)
vy = 15.0   # Kecepatan awal pada sumbu y (contoh nilai)
g = 9.8     # Percepatan gravitasi (m/s^2)
t = 0.0     # Waktu awal
dt = 0.1    # Perubahan waktu per langkah
x = 0.0     # Posisi awal pada sumbu x (contoh nilai)
y = 0.0     # Posisi awal pada sumbu y (contoh nilai)

# Inisialisasi list untuk menyimpan data
t_list = []
x_list = []
y_list = []
v_list = []


vy_t = vy - g * t
v_total = math.sqrt(vx**2 + vy_t**2)

t_list.append(t)
x_list.append(x)
y_list.append(max(0, y))
v_list.append(v_total)

t += dt

# Anda bisa mencetak beberapa hasil untuk verifikasi
print(f"vy_t saat ini: {vy_t}")
print(f"v_total saat ini: {v_total}")
print(f"Waktu yang diperbarui: {t}")
print(f"Daftar waktu: {t_list}")
```

```python
def hitung_energi(m, v, h, g=9.8):
      ek = 0.5 * m * v**2
      ep = m * g * h
      return ek, ep
```

```python
def validasi_input(prompt, tipe='float', min_val=None, max_val=None):
    while True:
        try:
            if tipe == 'float':
                nilai = float(input(prompt))
            elif tipe == 'int':
                nilai = int(input(prompt))
            else:
                print("Tipe input tidak didukung. Gunakan 'float' atau 'int'.")
                continue

            if min_val is not None and nilai < min_val:
                print(f"Nilai harus lebih besar atau sama dengan {min_val}.")
            elif max_val is not None and nilai > max_val:
                print(f"Nilai harus lebih kecil atau sama dengan {max_val}.")
            else:
                return nilai
        except ValueError:
            print("Input tidak valid. Masukkan angka yang benar.")
```
